```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
#plot X axis and Y axis in form of list-
plt.plot([1.5,2,3],[2,5.5,7])
plt.show()
```

```python
#label X,Y and chart data
plt.plot([1.5,2,3],[2,5.5,7])
plt.title("Bishal Data")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

```python
#two data in same plot
#new style ko plot named ggplot
style.use("ggplot")
x=[1.5,2,3]
y=[2,5.5,7]
x1=[1,2.2,2.9]
y1=[1.9,5.4,6.9]
#to show label of two data:
plt.plot(x,y,label="first Data",linewidth=3)
plt.plot(x1,y1,label="Second Data",linewidth=4)
plt.title("My Data")
plt.xlabel("x axis")
plt.ylabel("Y axis")
#to show the legend of two data
plt.legend()
plt.show()
```

```python
#plot in scattered way
plt.scatter(x,y,label="first Data",linewidth=3)
plt.scatter(x1,y1,label="Second Data",linewidth=4)
plt.title("My Data")
plt.xlabel("x axis")
plt.ylabel("Y axis")
#to show the legend of two data
plt.legend()
plt.show()
```
