```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Extract csv file and store it into df dataframe
#converting Date column data into time stamp and setting the date as index
df=pd.read_csv("missingdata.csv",parse_dates=["day"])
df
```

```python
#Setting index as Day
df.set_index("day",inplace=True)
df=df.interpolate()
plt.plot(df["day"],df["temperature"])
plt.show()
```

```python
#fillna function: fill any function into some value
df_new1=df.fillna(0)
```

```python
#But there is problem. Event value cannot be zero.
#We make temp and windspeed NANS Zero but put event NAns no event
df_new2=df.fillna({
    "temperature":0,
    "windspeed":0,
    "event":"no event"
})
```

```python
# zero halna could be unreasonable so . To keep NANs as adjecent neighbouring values:
#above data and limit only one abvoethe vlaue
df_new3=df.fillna(method="ffill",limit=1)
df_new3
#backward data
df_new4=df.fillna(method="bfill")
df_new4
#leftdata
df_new5=df.fillna(method="ffill",axis=1)
df_new5
#right data
df_new6=df.fillna(method="bfill",axis=1)

```

```python
#interpolate between two values:
df_new7=df.interpolate()
df
```

```python
# interpolate on the basis of time:
df_new8=df.interpolate(method="time")
df_new8
```

```python
#lets create delete entire row with atleast one NaN value
df_new9=df.dropna()
df_new9
#delete the row only when all elemnts have NaN value
df_new10=df.dropna(how="all")
df_new10
#keep the row if atleast N values 
df_new11=df.dropna(thresh=1)
df_new11
```
