```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style

#Read Data From Csv, convert csv into data frame and plot the data
df=pd.read_csv("plotdata.csv")
style.use("dark_background")
plt.plot(df["X"],df["Y"],label="First Data")
plt.plot(df["X1"],df["Y1"],label="Second Data")
plt.title("My Data")
plt.xlabel("X-Data")
plt.ylabel("Y-Data")
plt.legend()
plt.show()
```
